```python
from openai import OpenAI
import sqlite3
import pandas as pd
import os
```

```python
conn = sqlite3.connect('articles_database.db')

query = "SELECT * FROM articles"

# Execute the query and load data into a DataFrame
df = pd.read_sql_query(query, conn)

# Don't forget to close the connection when you're done
conn.close()
```

```python
df
```

```python
paper_text = df['Full Text'][0]
```

```python
metadata_features = [
    "number of participants",
    "study design"
]

def generate_analysis_prompts(paper_text, features):
    prompts = []
    for feature in features:
        prompt = f"Analyze the following scientific paper with a focus on its {feature}. Provide the {feature} based on the content of the paper using the format [{feature} = __ ]. You must report a confidence score in your answer, which is how confident you are that the returned result is the correct result. \n\n{paper_text}"
        prompts.append(prompt)
    return prompts

prompts = generate_analysis_prompts(paper_text, metadata_features)

print(prompts[0])
```

```python
with open('/workspaces/Wikistim-Summarization/my_GPT_key_2nd.txt', 'r') as file:
    API_KEY = file.read().strip()
```

```python
client = OpenAI(
    # This is the default and can be omitted
    api_key=API_KEY,
)


responses = []
for prompt in prompts:
    print(prompt)
    response = client.chat.completions.create(
        model="gpt-4-1106-preview",  # Adjust the model as necessary
        messages=[
            {"role": "user", "content": prompt}
        ],
        max_tokens=500
    )


```

```python
first_choice_content = response.choices[0].message.content.strip()
responses.append(first_choice_content)
```
